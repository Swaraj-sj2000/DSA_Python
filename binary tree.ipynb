```python
class Tree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def inputTree(self):
        leftData = int(input(f"Enter the left child of {self.data} (-1 for None): "))
        if leftData != -1:
            self.left = Tree(leftData)
            self.left.inputTree()

        rightData = int(input(f"Enter the right child of {self.data} (-1 for None): "))
        if rightData != -1:
            self.right = Tree(rightData)
            self.right.inputTree()

    

            
    def printTree(self):
        if self is None:
            return
        print(self.data, end=":")
        if self.left:
            print("L", self.left.data, end=",")
        if self.right:
            print("R", self.right.data)
        print()
        if self.left:
            self.left.printTree()
        if self.right:
            self.right.printTree()

    def countNodes(self):
        if self is None:
            return 0
        left_count = self.left.countNodes() if self.left else 0
        right_count = self.right.countNodes() if self.right else 0
        return 1 + left_count + right_count
    
    @staticmethod
    def inorderTraversal(root):
        l1=[]
        if root:
            l1+=Tree.inorderTraversal(root.left)
            l1.append(root.data)
            l1+=Tree.inorderTraversal(root.right)
        return l1
    
    @staticmethod
    def preorderTraversal(root):
        l1=[]
        if root:
            l1.append(root.data)
            l1+=Tree.preorderTraversal(root.left)
            l1+=Tree.preorderTraversal(root.right)
        return l1
        
    @staticmethod
    def postorderTraversal(root):
        l1=[]
        if root:
            l1+=Tree.postorderTraversal(root.left)
            l1+=Tree.postorderTraversal(root.right)
            l1.append(root.data)
            
        return l1
    
    
    @staticmethod
    def max_element(root):
        l_st=Tree.max_element(root.left) if root.left else -1
        r_st=Tree.max_element(root.right) if root.right else -1
        return max(root.data,max(l_st,r_st))
            

# Driver code
rootData = int(input("Enter the root data: "))
tree = Tree(rootData)
tree.inputTree()

tree.printTree()

print("Number of nodes:", tree.countNodes())

print(Tree.preorderTraversal(tree))
print(Tree.postorderTraversal(tree))
print(Tree.inorderTraversal(tree))
```
